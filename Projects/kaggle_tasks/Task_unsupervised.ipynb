```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_classification, make_blobs, make_moons

from sklearn.mixture import GaussianMixture
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
```

# Dataset 1 

```python
X1, y1 = make_classification(n_samples=1000, n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=1, random_state=10)
plt.scatter(X1[:, 0], X1[:, 1], s=10, c=y1);
```

```python
kmeans = KMeans(n_clusters=2, init='random', n_init=10)
kmeans.fit(X1)
y_kmeans_1 = kmeans.predict(X1)
plt.scatter(X1[:, 0], X1[:, 1], s=10, c=y_kmeans_1);
```

```python
model = AgglomerativeClustering(n_clusters=None, linkage='average', distance_threshold=3)
p = model.fit_predict(X1)
plt.scatter(X1[:, 0], X1[:, 1], s=10, c=p);
```

```python
model = AgglomerativeClustering(n_clusters=2, linkage='ward', distance_threshold=None, compute_full_tree='auto')
p1 = model.fit_predict(X1)
plt.scatter(X1[:, 0], X1[:, 1], s=10, c=p1);
```

```python
gmm_sk = GaussianMixture(n_components=2)
gmm_sk.fit(X1)
preds_sk = gmm_sk.predict(X1)
plt.scatter(X1[:, 0], X1[:, 1], s=10, c=preds_sk);
```

```python
model_db = DBSCAN(eps=0.51, min_samples=150)
y_pred_db = model_db.fit_predict(X1)
plt.scatter(X1[:, 0], X1[:, 1], s=10, c=y_pred_db);
```

# Dataset 4

```python
rng = np.random.RandomState(13)
X4 = np.dot(X2, rng.randn(2, 2))

plt.scatter(X4[:, 0], X4[:, 1], s=30);
```

```python
model = AgglomerativeClustering(n_clusters=4, linkage='ward', distance_threshold=None, compute_full_tree='auto')
p1 = model.fit_predict(X4)
plt.scatter(X4[:, 0], X4[:, 1], s=10, c=p1);
```

```python
model = AgglomerativeClustering(n_clusters=None, linkage='average', distance_threshold=3)
p = model.fit_predict(X4)
plt.scatter(X4[:, 0], X4[:, 1], s=10, c=p);
```

```python
kmeans = KMeans(n_clusters=4, init='random', n_init=10)
kmeans.fit(X4)
y_kmeans_2 = kmeans.predict(X4)
plt.scatter(X4[:, 0], X4[:, 1], s=10, c=y_kmeans_2);
```

```python
model_db = DBSCAN(eps=0.3, min_samples=3)
y_pred_db = model_db.fit_predict(X4)
plt.scatter(X4[:, 0], X4[:, 1], s=10, c=y_pred_db);
```

```python
gmm_sk = GaussianMixture(n_components=3)
gmm_sk.fit(X4)
preds_sk = gmm_sk.predict(X4)
plt.scatter(X4[:, 0], X4[:, 1], s=10, c=preds_sk);
```

```python
k = list(range(2, 20))
l = []
for i in k:
    kmeans = KMeans(n_clusters=i, init='random', n_init=10)
    kmeans.fit(X4)
    l.append(abs(kmeans.score(X4)))

plt.plot(k, l)
plt.grid()
```
